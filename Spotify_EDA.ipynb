```python
#understanding error bar
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [10, 15, 7, 12]
errors = [1, 2, 0.5, 1]

# Create a basic plot with error bars
plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Error Bars')

# Show the plot
plt.show()
```
